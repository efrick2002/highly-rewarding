```python
df = df.merge(hp[['question_id', 'prompt', 'response_1', 'response_2']], on="question_id")
df
```

```python
df['var_delta'] = np.abs(df.var_1 - df.var_2)
df['score_delta'] = np.abs(df.score_1 - df.score_2)

df['var_max'] = np.maximum(df.var_1, df.var_2)
df['var_min'] = np.minimum(df.var_1, df.var_2)


df.plot.scatter(x="score_delta", y="var_delta")

```

```python
df = df.sort_values(by="var_delta", ascending=False)
df[["prompt", "response_1", "response_2", "score_1", "score_2", "var_1", "var_2", "score_delta", "var_delta", "var_max", "var_min"]].set_index(["prompt", "response_1", "response_2"]).head(10)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Required for 3D plotting
from scipy.stats import gaussian_kde

# Create a grid over which to evaluate the KDE.
x_min, x_max = df['score_1'].quantile(0.01), df['score_1'].quantile(0.99)
y_min, y_max = 0.0, df['var_1'].quantile(0.99)

# You can adjust the resolution by changing the number of points
x_grid = np.linspace(x_min, x_max, 100)
y_grid = np.linspace(y_min, y_max, 100)
X, Y = np.meshgrid(x_grid, y_grid)

# Stack the grid positions for KDE evaluation
positions = np.vstack([X.ravel(), Y.ravel()])
values = np.vstack([df['score_1'], df['var_1']])
kde = gaussian_kde(values)
Z = np.reshape(kde(positions), X.shape)

# Plot the smooth surface
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Add labels and title
ax.set_xlabel('score_1')
ax.set_ylabel('var_1')
ax.set_zlabel('Density')
ax.set_title('Smooth 3D KDE Surface for score_1 vs var_1')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

plt.show()



```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from scipy.stats import gaussian_kde

# Create a grid over which to evaluate the KDE.
x_min, x_max = df['score_1'].quantile(0.05), df['score_1'].quantile(0.99)
y_min, y_max = 0.0, df['var_1'].quantile(0.95)
x_grid = np.linspace(x_min, x_max, 100)
y_grid = np.linspace(y_min, y_max, 100)
X, Y = np.meshgrid(x_grid, y_grid)

# Prepare the data for KDE.
positions = np.vstack([X.ravel(), Y.ravel()])
values = np.vstack([df['score_1'], df['var_1']])
kde = gaussian_kde(values)
Z = np.reshape(kde(positions), X.shape)

# Create an interactive 3D surface plot using Plotly.
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale='Viridis')])
fig.update_layout(
    title='Interactive 3D KDE Surface Plot',
    scene=dict(
        xaxis_title='score_1',
        yaxis_title='var_1',
        zaxis_title='Density'
    )
)
fig.show()
```
